```python
def getPriceSpread( position: int, stage:tuple[int], spread:tuple[tuple[int, int]]) -> tuple[int, int]:
    sign = 1 if position >= 0 else -1
    for i, pos in enumerate(stage):
        if abs(position) <= pos:
            return spread[i][::sign]
```
